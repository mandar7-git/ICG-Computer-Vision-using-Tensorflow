```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#linear plot
x= np.arange(1,8)
y = [14,8,6,12,10,4,8]

plt.xlabel("X axis", size = 15)
plt.ylabel("Y axis", size = 15)
plt.plot(x,y)
plt.show()

#scatter plot
plt.xlabel("X axis", size = 15)
plt.ylabel("Y axis", size = 15)
plt.scatter(x,y)
plt.show()

#bar graph
plt.xlabel("X axis", size = 15)
plt.ylabel("Y axis", size = 15)

plt.bar(x,y)
plt.show()



```
