## Model Definition

Using TensorFlow, build a model with the following definition:
> Input of shape 5 \\
> Dense of shape 5 \\
> Dense of shape 5 \\
> Dense of shape 1 \\

Use Mean Square Error Loss and Stochaistic Gradient Descent (SGD) Optimizer

Use Gradient Decay with appropriate parameters

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential()
model.add(keras.Input(shape=(5)))
model.add(layers.Dense(5, activation='relu'))
model.add(layers.Dense(5, activation='relu'))
model.add(layers.Dense(1))
# compile model
optimizer = keras.optimizers.SGD(learning_rate=0.1)
model.compile(loss='mse')
# fit model
history = model.fit(X_train, y_train, validation_split=0.33, epochs=20,batch_size=1)
```

## Plot Loss

Using matplotlib visualise how the loss (both validation and training) is changing, use this information to retrain the model with appropriate parameters.<br>We ideally want the loss to be constant over the last few iterations.

```python
# Your code here
import matplotlib.pyplot as plt
loss_train = history.history['loss']
loss_val = history.history['val_loss']
plt.plot(loss_train, 'b', label='Training loss')
plt.plot(loss_val, 'g', label='validation loss')
plt.title('Training and Validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

## Evaluation Metrics
Use the R2 Score function implemented in the first assignment to evaluate the performance of the model.

```python
import numpy as np
predictions=model.predict(X_test)
from sklearn.metrics import r2_score
accu = r2_score(y_test, predictions) 
print("Accuracy:", accu)
```

# Generating Dataset 

```python
# Your code here
from sklearn import datasets
from sklearn.model_selection import train_test_split

# Generate the data
X, y = datasets.make_regression(n_samples=10000, n_features=5, noise=5, random_state=4)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

# Visualizing the Dataset

```python
import matplotlib.pyplot as plt
for i in range(0,5):
  plt.scatter(X[:,i],y)
  plt.show()
```

# Model Definition

Using TensorFlow,we have built a model with the following definition:



Input of shape 5

Dense of shape 6

Dense of shape 6

Dense of shape 6

Dense of shape 4

Dense of shape 1

We have used Mean Square Error Loss and Adam Optimizer











```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential()
model.add(keras.Input(shape=(5)))
model.add(layers.Dense(6, activation='relu'))
model.add(layers.Dense(6, activation='relu'))
model.add(layers.Dense(6, activation='relu'))
model.add(layers.Dense(4, activation='relu'))
model.add(layers.Dense(1))
# compile model
optimizer= keras.optimizers.Adam(learning_rate=0.0003)
model.compile(loss='mse', optimizer= optimizer)

# fit model
history = model.fit(X_train, y_train, validation_split=0.33, epochs=10,batch_size=1)
```

# Plotting the Loss

```python
import matplotlib.pyplot as plt
loss_train = history.history['loss']
loss_val = history.history['val_loss']
plt.plot(loss_train, 'g', label='Training loss')
plt.plot(loss_val, 'b', label='validation loss')
plt.title('Training and Validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

# Evaluation Metrics

Calculating the r2_score and accuracy of our model

```python
import numpy as np
predictions=model.predict(X_test)
from sklearn.metrics import r2_score
accu = r2_score(y_test, predictions) 
print("Accuracy:", accu)
```
